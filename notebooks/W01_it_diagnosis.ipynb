## Step 0: 환경 설정

```python
import subprocess
subprocess.run(['apt-get','-qq','-y','install','fonts-nanum'],capture_output=True)
!pip install -q google-generativeai pandas matplotlib
import matplotlib.pyplot as plt, matplotlib.font_manager as fm
_n=[f for f in fm.findSystemFonts() if 'Nanum' in f]
if _n: fm.fontManager.addfont(_n[0]); plt.rcParams['font.family']=fm.FontProperties(fname=_n[0]).get_name()
plt.rcParams['axes.unicode_minus']=False
try:
    from google.colab import userdata; API_KEY=userdata.get('GEMINI_API_KEY')
except: API_KEY=input('GEMINI_API_KEY: ')
import google.generativeai as genai; genai.configure(api_key=API_KEY)
model=genai.GenerativeModel('gemini-2.0-flash'); print('✅ 준비 완료')
```

## Step 1: 기관 정보 입력 (✏️)

```python
INFO={
    'name':'✏️ [교육 기업명]',
    'type':'✏️ [SI|SaaS|컨설팅|스타트업|에이전시]',
    'employees':'✏️ [직원수]',
    'main_service':'✏️ [주요 서비스]',
}
print('✅',INFO['name'])
```

## Step 2: AX 진단 문항 (각 1~5점 입력)

```python
SCORES={
    '제안서/기술사양서 자동화':1, # ✏️
    '코드 리뷰 자동화':1,         # ✏️
    '프로젝트KPI 분석':1,         # ✏️
    '버그/이슈 리포트 자동화':1,  # ✏️
    '고객 커뮤니케이션 자동화':1, # ✏️
    '배포/CI-CD 자동화':1,        # ✏️
    '보안 체크 자동화':1,         # ✏️
    '품질 테스트 자동화':1,       # ✏️
    '개발 문서화 자동화':1,       # ✏️
    '통합 IT 대시보드':1,         # ✏️
}
total=sum(SCORES.values())
level='🟢 Starter' if total<25 else ('🔵 Growing' if total<40 else '🟣 AI-Ready')
print(f'총점: {total}/50 — {level}')
```

## Step 3: 시각화

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(10,5))
ax.barh(list(SCORES.keys()),list(SCORES.values()),
        color=['#e74c3c' if v<=2 else '#f39c12' if v<=3 else '#2ecc71' for v in SCORES.values()])
ax.axvline(3,color='gray',linestyle='--',alpha=0.5,label='기준(3점)')
ax.set_xlim(0,5); ax.legend()
ax.set_title(f'{INFO["name"]} AX 진단 — {level}')
plt.tight_layout()
plt.savefig('ax_diag.png',dpi=150,bbox_inches='tight'); plt.show()
```

## Step 4: Gemini 맞춤 전략 리포트

```python
weak=[k for k,v in SCORES.items() if v<=2]
prompt=f"""IT서비스업 AI 전환 전문 컨설턴트로서:
기관: {{INFO['name']}} ({{INFO['type']}})
총점: {{total}}/50점 — {{level}}
즉시 개선 항목: {{', '.join(weak) if weak else '전반적 고도화'}}
12주 로드맵, 즉시 실행 3가지, 기대 ROI 포함. 마크다운 형식."""
resp=model.generate_content(prompt)
print(resp.text)
```

## Step 5: 다운로드

```python
with open('ax_report.md','w',encoding='utf-8') as f:
    f.write(f'# {{INFO["name"]}} AX 진단 리포트\n\n총점: {{total}}/50 | {{level}}\n\n')
    f.write(resp.text)
from google.colab import files
files.download('ax_report.md'); files.download('ax_diag.png')
print('✅ 완료!')
```
