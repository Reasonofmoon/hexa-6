## Step 2: 스프린트/버그/배포/SLA 4패널 대시보드+Gemini 인사이트

```python
import pandas as pd, numpy as np
np.random.seed(42)
# 스프린트별 샘플 데이터
sprints=pd.DataFrame({'스프린트':[f'S{i+1}' for i in range(10)],
    '완료율':np.random.uniform(60,100,10).round(1),
    '버그수':np.random.randint(0,20,10),
    '배포횟수':np.random.randint(1,5,10)})
# 이상치 삽입 (W09용)
sprints.loc[3,'버그수']=85; sprints.loc[7,'버그수']=92
print(sprints.to_string(index=False))
```

## Step 3: AI 분석

```python
p=f"""IT 서비스 AI 전화 컨설턴트로서 스프린트/버그/배포/SLA 4패널 대시보드+Gemini 인사이트를 수행하세요.
기업: {{INFO['name']}} ({{INFO['type']}}), 프로젝트: {{INFO['project']}}
인사이트 3가지 + 즉시 액션. 마크다운."""
resp=model.generate_content(p); print(resp.text)
```
